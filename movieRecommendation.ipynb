```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data=pd.read_csv('/content/movies.csv')
```

```python
movies_data.shape
```

selecting features

```python
movies_data.head()
```

```python
selected=['genres',"keywords","tagline","cast","director","original_title"]

for feature in selected:
  movies_data[feature]=movies_data[feature].fillna('')
```

```python
combined_features=movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']+' '+movies_data["original_title"]
```

```python
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
similarity=cosine_similarity(feature_vectors)

print(similarity)
```

```python
user_input=input("Movie Name : ")

all_titles=movies_data['title'].tolist()

close_match=difflib.get_close_matches(user_input,all_titles)
closest=close_match[0]

index_of_movie=movies_data[movies_data.title == closest]['index'].values[0]

similarity_score = [(i, float(score)) for i, score in enumerate(similarity[index_of_movie])]

sorted_similarity_score=sorted(similarity_score,key=lambda x:x[1],reverse=True)

print("Suggested Movies : \n")

i=1

for movie in sorted_similarity_score:
  index=movie[0]
  title_of_movie=movies_data[movies_data.index==index]['title'].values[0]

  if (i<30):
    print(i,'.',title_of_movie)
    i+=1
```
